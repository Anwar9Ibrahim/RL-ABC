# Visualization of ACC Elegant RL Training Results
This notebook demonstrates how to use the `ElegantVisualizer` class to visualize the results of the Elegant simulations.

```python
import os
import sys
import matplotlib.pyplot as plt
from IPython.display import display, HTML

# Import configuration and environment setup
from config_manager import load_config
from train import RLConfig
from rl_framework.visulize import ElegantVisualizer
```

## Environment Setup
First, we load the configuration and initialize the environment.

```python
# Load configuration from config.yaml || configs.json
config_dict = load_config('config.json')

# Initialize configuration object   
config = RLConfig(config_dict)

# Override init_num_particles if needed
config.init_num_particles = 1000

# Setup the environment
env, logger, file_handler = config.setup_environment()
```

## Define Magnet Values
We define a dictionary of magnet values (quadrupoles and bending magnets) that we want to visualize.

```python
# Values to try on the environment
values = {
    'Q1L0K1': 11.1414, 'Q1L0HKICK': 0.005, 'Q1L0VKICK': 0.005, 
    'Q1L1K1': -11.0152, 'Q1L1HKICK': -0.005, 'Q1L1VKICK': -0.0023, 
    'BM1FSE': 0.005, 
    'Q1L2_1K1': 10.1493, 'Q1L2_1HKICK': 0.0023, 'Q1L2_1VKICK': -0.005, 
    'Q1L3_1K1': -5.1645, 'Q1L3_1HKICK': 0.0023, 'Q1L3_1VKICK': 0.0026, 
    'Q1L4_1K1': 7.9541, 'Q1L4_1HKICK': -0.0021, 'Q1L4_1VKICK': -0.005, 
    'Q1L3_2K1': -5.9545, 'Q1L3_2HKICK': 0.004, 'Q1L3_2VKICK': 0.0027, 
    'Q1L2_2K1': 10.2039, 'Q1L2_2HKICK': -0.0015, 'Q1L2_2VKICK': -0.004, 
    'BM4FSE': 0.005, 
    'Q1L7K1': -12.3519, 'Q1L7HKICK': -0.0027, 'Q1L7VKICK': -0.0005, 
    'Q1L4_2K1': 13.9352, 'Q1L4_2HKICK': 0.0016, 'Q1L4_2VKICK': 0.0016, 
    'Q1L9K1': -9.1579, 'Q1L9HKICK': -0.0031, 'Q1L9VKICK': -0.0009, 
    'Q1L10K1': 13.8398, 'Q1L10HKICK': -0.0048, 'Q1L10VKICK': -0.005
}

# Initialize the visualizer
visualizer = ElegantVisualizer(env, values)
```

## 1. Plot SDDS Data
Plot the phase space data from an SDDS file.

```python
# Make sure the results directory exists and the simulation has generated the files
# We can use WQ1L1_1.sdds or final_WP.sdds depending on what is generated
sdds_file = "results_0/WQ1L1_1.sdds"

if os.path.exists(sdds_file):
    # Plot raw SDDS data
    img = visualizer.plot_sdds_data(sdds_file, png_path=f"results_0/image.png")
    display(img)
else:
    print(f"File {sdds_file} not found. Make sure that you used the correct path and that the simulation ran successfully.")
```

## 2. Plot Edited SDDS Data
Plot the processed SDDS data (e.g., converted to mm and mrad).

```python
if os.path.exists(sdds_file):
    edited_img = visualizer.plot_sdds_data_edited(sdds_file, png_path=f"results_0/image_edited.png")
    display(edited_img)
```

## 3. Plot Energy Phase Space
Plot energy vs phase space coordinates.

```python
if os.path.exists(sdds_file):
    energy_img = visualizer.plot_energy_phase_space(sdds_file, png_path=f"results_0/energy_phase_space.png")
    display(energy_img)
```

## 4. Plot Magnets
Visualize the magnet layout.

```python
mags, htmlmags = visualizer.plot_magnets()
#display(mags)
display(htmlmags)
```

Visualize the tune diagram.

```python

from PIL import Image as PILImage

try:
    image, res_1 = visualizer.plot_tune_diagram()
    display(image)
    img=PILImage.open('results_0/image.png')
    display(img)
    display(res_1)
except Exception as e:
    print(f"Error plotting tune diagram: {e}")
```
